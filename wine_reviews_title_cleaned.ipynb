```python
import pandas as pd
```

```python
df = pd.read_csv("intermediate_data/wine_reviews.csv")
df.head()
```

```python
df_clean = pd.read_csv("intermediate_data/wine_reviews_title_cleaned.csv")
df_clean.head()
```

```python
pd.DataFrame().assign(title_old=df.title.sort_values(), title_clean=df_clean.title.sort_values())
```

## Cleaning steps in OpenRefine

* Trim leading and trailing whitespaces
* Collapse consecutive whitespaces
* Cluster and remove redundant title values (374 cells merged)
* Remove redundant location information by running `value.replace(/\W+\(.+\)$/, '')`
* Re-clustering to further remove redundant titles (68 cells merged)

```python
df_title_unique = df.title.unique().size
df_clean_title_unique = df_clean.title.unique().size
reduction_rate = (df_title_unique - df_clean_title_unique) / df_title_unique
print(f'wine_reviews had {df_title_unique} unique titles')
print(f'wine_reviews_title_cleaned had {df_clean.title.unique().size} unique titles')
print(f'{reduction_rate*100:.2f}% defect reduction')
```
